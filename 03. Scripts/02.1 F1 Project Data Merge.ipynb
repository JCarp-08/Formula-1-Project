## 03. Data Merger

```python
#merging driverID on results
dr_merge = drid.merge(res, on = ['driverId','number'])
#removing unwanted column
dr_merge = dr_merge.drop('dob', axis = 1)
dr_merge
```

```python
dr_merge.columns
```

```python
#renaming column to resolve potential conflict
cnid = cnid.rename(columns = {'nationality':'con_nat'})
cnid.head()
```

```python
#exporting change
cnid.to_pickle(r'C:\Users\jmanc\OneDrive\Desktop\F1 Project\Data\Prepped Data\constructorID.pkl')
```

```python
#merging constructorID to dr_merge
dr_merge = dr_merge.merge(cnid, on = 'constructorId')
dr_merge
```

```python
#creating merged set for circuits and races
rcs_merge = rcs.merge(crcid, on = ['circuitId'])
rcs_merge
```

```python
#renaming columns
rcs_merge = rcs_merge.rename(columns = {'name_x':'gp_name', 'name_y':'circ_name'})
```
